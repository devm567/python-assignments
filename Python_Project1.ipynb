```python
import pandas as pd
```

```python
import numpy as np
```

```python
data = pd.read_csv("Travel.csv")
```

```python
1. What is the distribution of customerages, and how does age relate to product purchase (ProdTaken)?
```

```python
data_prod_taken_1 = data[data['ProdTaken'] == 1]
```

```python
age = data_prod_taken_1['Age'].value_counts()
print(age)
```

```python
pt = pd.pivot_table(data, index='Age', values='ProdTaken', aggfunc=np.sum)
```

```python
pt
```

```python
2. How does the type of contact(type of contact)influence the likelihood of purchasing a
travel product?  
```

```python
typeofcontact=pd.pivot_table(data, values='ProdTaken', index='TypeofContact', aggfunc=np.sum)
```

```python
typeofcontact
```

```python
3. Is there a correlation between the city tier(citytier)and monthly income (monthly
income) of customers?
```

```python
ctmi = pd.pivot_table(data, values='MonthlyIncome', index='CityTier', aggfunc=np.mean)
```

```python
ctmi
```

```python
4 . What is the average duration of a pitch (duration of pitch)for customers who
purchased a product versus those who did not? 
```

```python
purchased =  pd.pivot_table(data_prod_taken_1 , values='DurationOfPitch', index='ProductPitched', aggfunc=np.mean)
purchased
```

```python
data_prod_taken_0 = data[data['ProdTaken']==0]
```

```python
notpurchased =  pd.pivot_table(data_prod_taken_0 , values='DurationOfPitch', index='ProductPitched', aggfunc=np.mean)
notpurchased
```

```python
5. How do occupation types(occupation)distribute among customers,and is there an
occupation type that is more likely to purchase a travel product?
```

```python
occupation=data['Occupation'].value_counts()
occupation
```

```python
6. Are there any gender-based preferences for the type of product pitched
(ProductPitched)?
```

```python
ch = data.groupby("Gender")["ProductPitched"].count()
```

```python
ch
```

```python
7. What is the relationship between the number of trips(number of trips)taken by a
customer and their likelihood to purchase a new travel product?
```

```python
relation=pd.pivot_table(data,values='ProdTaken',index='NumberOfTrips',aggfunc=np.count_nonzero)
relation
```

```python
8. Does having a passport(passport)correlate with a higher number of trips taken or a
higher likelihood of purchasing a travel product? 
```

```python
highernum = data.groupby('Passport')['NumberOfTrips'].sum()
highernum
```

```python
9. How satisfied are customers with the pitch (pitch satisfaction score),and does this
satisfaction influence product purchase? 
```

```python
satisfaction=data.groupby('PitchSatisfactionScore')['CustomerID'].count()
satisfaction
```

```python
10. Among customers with children (number of children visiting), how does the number
of children impact travel product purchases? 
```

```python
impact = pd.pivot_table(data, index='NumberOfChildrenVisiting', values= 'ProdTaken', aggfunc=np.sum)
```

```python
impact
```

```python
11. Gender preferences for product types: how does gender influence the preference for
different types of travel products pitched (product pitched)? 
```

```python
product = pd.pivot_table(data, index='Gender', columns='ProductPitched', aggfunc='size', fill_value=0)
product
```
